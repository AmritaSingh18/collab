```python
from datasets import load_dataset
multinews = load_dataset("multi_news",split='test', cache_dir="./cache")
```

```python
multinews.to_pandas()
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("t5-small")
```

```python
multi_news = multinews.train_test_split(test_size=0.2)
```

```python
def process_function(examples):
  prefix = "summarize: "
  inputs=[prefix + doc for doc in examples['document']]
  model_inputs = tokenizer(inputs, max_length=1024, truncation=True)
  labels = tokenizer(examples['summary'], max_length=128, truncation=True)
  model_inputs['labels']=labels['input_ids']
  return model_inputs
```

```python
tokenized_multi_news = multi_news.map(process_function, batched=True)
```

```python
from transformers import DataCollatorForSeq2Seq, AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=10,
    predict_with_generate=True,
    fp16=True,
    push_to_hub=False,
    report_to = "none"
)
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_multi_news["train"],
    eval_dataset=tokenized_multi_news["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
)
```

```python
trainer.train()
```

```python
text= """ Self Attention models
Introduction:
In the ever-evolving landscape of artificial intelligence, the emergence of self-attention models has been nothing short of revolutionary. These models have transformed the way machines understand and generate content, offering unparalleled capabilities in capturing intricate patterns and dependencies within data. In this article, we'll delve into the intricacies of self-attention models in Generative AI, exploring their principles, techniques, and showcasing their prowess through an illustrative example.

Understanding Self-Attention:
At its core, self-attention is a mechanism that allows neural networks to weigh the importance of different parts of an input sequence when generating an output sequence. It enables the model to focus on relevant information while suppressing noise or irrelevant details, leading to more precise and context-aware predictions.

Self-Attention Model:
A self-attention model consists of three essential components:

Input Embeddings: The input sequence, whether it's text, images, or any other form of data, is first transformed into a set of embeddings that capture its semantic meaning and relationships.

Attention Mechanism: This is the heart of the self-attention model. It computes attention scores for each element in the input sequence, determining how much focus to place on each element when generating the output sequence. These attention scores are computed based on the similarity between elements in the input sequence.

Output Generation: Finally, the model combines the input embeddings with the attention scores to generate the output sequence. By weighting the contributions of different input elements based on their relevance, the model produces more contextually rich and accurate outputs.

Example: Text Summarization
Let's illustrate the power of self-attention models through an example of text summarization. Suppose we have a lengthy article about artificial intelligence, and we want to generate a concise summary of its key points.

Input Embeddings: We first encode each word in the article into a vector representation using techniques like word embeddings (e.g., Word2Vec, GloVe).

Attention Mechanism: The self-attention mechanism computes attention scores for each word in the article, determining its importance in the context of summarization. Words that are more relevant to the main ideas of the article are assigned higher attention scores.

Output Generation: The model combines the input embeddings with the attention scores to generate the summary. It selects words with higher attention scores, ensuring that the summary captures the essence of the original article while being concise and informative."""
```
